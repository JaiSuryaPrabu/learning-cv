# Image annotation

In this notebook, I will learn about :
1. Draw line
2. Draw circle
3. Draw rectangle
4. Write a text in image

## Importing libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Drawing a line

To draw a line we use `cv2.line()` function and the arguments are :
1. img - image data
2. pt1 - starting point (x,y) in tuple
3. pt2 - ending point (x,y) in tuple
4. color - color of the line
Optional arguments :
1. thickness - thickness of the line, by default it is 1
2. lineType - default is 8 and it is `cv2.LINE_AA` (antialiased or smooth line)

```python
image = cv2.imread('images/car.jpg',cv2.IMREAD_GRAYSCALE)
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
imageLine = image_rgb.copy()
imageLine = cv2.line(imageLine,(100,250),(400,250),color=(0,255,255),thickness=2)
plt.imshow(imageLine[...,::-1])
```

## Drawing a circle

To draw a circle we use `cv2.circle()` function and the following arguments are:
1. image - image data
2. center - center point in tuple (x,y)
3. radius - radius value
4. color - tuple value of bgr value
5. lineType - same as in line

```python
imageCircle = image_rgb.copy()

imageCircle = cv2.circle(imageCircle,center = (250,150), radius = 50, color = (0,255,255),thickness=1, lineType = cv2.LINE_AA)
plt.imshow(imageCircle[...,::-1])
```

## Drawing a rectangle

To draw a rectangle we use `cv2.rectangle()` and the arguments are :
1. img - image data
2. pt1 - top left vertex
3. pt2 - bottom right vertex
4. color - bgr color value
5. thickness - if thickness is negative then it fills the entire rectangle with the color
6. lineType

```python
imageRec = image_rgb.copy()
imageRec = cv2.rectangle(imageRec,pt1=(120,120),pt2=(360,240),color=(0,255,255),thickness = -1,lineType=cv2.LINE_AA)
plt.imshow(imageRec[...,::-1])
```

## Write a text

To write a text in an image we use `cv2.putText()` method and the arguments are :
1. img - image data
2. text - text string to be written
3. org - bottom - left corner to where the text needs to be present
4. fontFace - Font type
5. fontScale - Font face factor
6. color - font color
7. thickness - by default 1
8. lineType - by default 8 or `cv2.LINE_AA`

```python
imageText = image_rgb.copy()
imageText = cv2.putText(imageText,text = "SPORTS CAR IMAGE",org = (150,280),fontFace=cv2.FONT_HERSHEY_PLAIN,fontScale = 1.2, color = (0,255,255),thickness=2)
plt.imshow(imageText[...,::-1])
```
