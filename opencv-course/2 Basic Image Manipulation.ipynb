# Basic Image Manipulation

In this notebook I will learn about :
* Accessing and managing imaage pixels
* Image resizing
* Cropping
* Flipping

## Importing Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Accessing and managing image pixels

```python
checkbox_image = cv2.imread(filename = "images/checkerboard_18x18.png", flags = 0)
print(checkbox_image)
plt.imshow(checkbox_image,cmap="gray")
```

```python
# accessing individual pixel
checkbox_image[0,0],checkbox_image[0,7]
```

```python
# modifying the pixels
checkbox_copy = checkbox_image.copy()
checkbox_copy[2,2] = 200
checkbox_copy[2,3] = 200
checkbox_copy[3,2] = 200
checkbox_copy[3,3] = 200

print(checkbox_copy)
plt.imshow(checkbox_copy,cmap="gray")
```

## Cropping Image

```python
filename = "images/New_Zealand_Lake.jpg"
lake_bgr = cv2.imread(filename = filename, flags = cv2.IMREAD_COLOR)
lake_rgb = cv2.cvtColor(lake_bgr,cv2.COLOR_BGR2RGB)
plt.imshow(lake_rgb)
```

```python
cropped_region = lake_rgb[200:400,300:500]
plt.imshow(cropped_region)
print(cropped_region.shape)
```

## Resize the image

To resize the image we use `cv2.resize()` function.
Arguments :
* src - source data
* dsize - destination size
* Optional :
    * fx - scale factor for horizontal axis
    * fy - scale factor for vertical axis


### 1. Specifying the scale factor x and y

```python
cropped_region_2x = cv2.resize(cropped_region,dsize=None,fx=2,fy=2)

plt.imshow(cropped_region_2x)
print(cropped_region_2x.shape)
```

### 2. Specifying the exact size of the output

```python
d_width = 100
d_height = 200
dim = (d_width,d_height)

r_cropped_region = cv2.resize(cropped_region,dsize=dim,interpolation=cv2.INTER_AREA)
plt.imshow(r_cropped_region)
```

## Flipping Images

Flipping the image is done by `cv2.flip()` function
Arguments :
* src - source image data
* flipCode - 1 means horizontal, 0 means vertical and -1 means both horizontal and vertical axis

```python
hori = cv2.flip(cropped_region,flipCode = 1)
vert = cv2.flip(cropped_region,flipCode = 0)
both = cv2.flip(cropped_region,flipCode = -1)

plt.figure(figsize=(20,5))
plt.subplot(141);plt.imshow(hori);plt.title("Horizontal flip")
plt.subplot(142);plt.imshow(vert);plt.title("Vertical flip")
plt.subplot(143);plt.imshow(both);plt.title("Both axis flip")
plt.subplot(144);plt.imshow(cropped_region);plt.title("Original")
```
